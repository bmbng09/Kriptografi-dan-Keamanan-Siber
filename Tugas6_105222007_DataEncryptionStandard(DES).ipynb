#### Helper Func

```python
def permute(block, table):
    return [block[i - 1] for i in table]

def xor(a, b):
    return [i ^ j for i, j in zip(a, b)]

def string_to_bit_array(s):
    return [int(bit) for char in s for bit in f'{ord(char):08b}']

def bit_array_to_string(arr):
    return ''.join(chr(int(''.join(map(str, arr[i:i+8])), 2)) for i in range(0, len(arr), 8))

def split_list(lst, n):
    return lst[:n], lst[n:]

def sbox_substitution(block):
    result = []
    for i in range(8):
        chunk = block[i*6:(i+1)*6]
        row = (chunk[0] << 1) | chunk[5]
        col = (chunk[1] << 3) | (chunk[2] << 2) | (chunk[3] << 1) | chunk[4]
        val = SBOXES[i][row % 4][col % 4]
        bin_val = [int(x) for x in f"{val:04b}"]
        result += bin_val
    return result

def feistel_function(r, k):
    print("  Feistel input R:", ''.join(map(str, r)))
    r_expanded = permute(r, E)
    print("  After Expansion:", ''.join(map(str, r_expanded)))
    r_xored = xor(r_expanded, k)
    print("  After XOR with Key:", ''.join(map(str, r_xored)))
    sbox_output = sbox_substitution(r_xored)
    print("  After S-Box Substitution:", ''.join(map(str, sbox_output)))
    permuted = permute(sbox_output, P)
    print("  After P Permutation:", ''.join(map(str, permuted)))
    return permuted

def generate_round_keys(key):
    key_bits = string_to_bit_array(key)
    return [key_bits[:48]] * 16

def des_encrypt_block(plaintext_block, key_block):
    print("\n--- DES Encryption Start ---")
    print("Plaintext:", plaintext_block)
    block = string_to_bit_array(plaintext_block)
    print("Binary Plaintext:", ''.join(map(str, block)))
    block = permute(block, IP)
    print("After Initial Permutation:", ''.join(map(str, block)))
    L, R = split_list(block, 32)

    round_keys = generate_round_keys(key_block)

    for i in range(16):
        print(f"\n-- Round {i+1} --")
        temp = R
        f_output = feistel_function(R, round_keys[i])
        R = xor(L, f_output)
        L = temp
        print("  L:", ''.join(map(str, L)))
        print("  R:", ''.join(map(str, R)))

    combined = R + L
    print("\nBefore Final Permutation:", ''.join(map(str, combined)))
    ciphertext_bits = permute(combined, FP)
    print("After Final Permutation:", ''.join(map(str, ciphertext_bits)))
    print("--- DES Encryption End ---\n")
    return ciphertext_bits

```
